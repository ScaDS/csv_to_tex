# Curate and Convert Lectures Notebook

This notebook processes a CSV file containing lecture data. The steps include loading the data, cleaning it, filtering it, and finally converting it to a LaTeX formatted string for document preparation.

```python
import pandas as pd
```

```python
filename = 'lectures_seminars-example.csv'
```

## Load the CSV file into a dataframe

```python
df = pd.read_csv(filename, header=1)

df.head(10)
```

## Replace non-numeric strings in `Num_students` with zero

```python
df['Num_students'] = pd.to_numeric(df['Num_students'], errors='coerce').fillna(0)

df.head(10)
```

## Convert `Num_students` to integers

```python
df['Num_students'] = df['Num_students'].astype(int)

df.head(10)
```

## Select specific columns

```python
df = df[["Semester", "Titel", "Num_students", "Betreuer_aus_scadsai", "Uni"]]
df.head(10)
```

## Filter out empty lines

```python
df = df.dropna(how='all')
df.head(10)
```

## Filter rows containing '2023' in the 'Semester' column

```python
df = df[df['Semester'].str.contains("2023", na=False)]
df.head(10)
```

## Filter rows where `Num_students` > 5

```python
df = df[df['Num_students'] > 5]

df.head(10)
```

## Convert the dataframe to a LaTeX formatted string and save to a .tex file

```python
latex_string = df.to_latex()

with open(filename.replace(".csv", ".tex"), 'w') as f:
    f.write(latex_string)
```

## Print the LaTeX formatted string

```python
print(latex_string)
```
